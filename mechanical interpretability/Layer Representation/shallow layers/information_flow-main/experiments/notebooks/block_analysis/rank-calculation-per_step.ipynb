```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import os

json_paths = {
    '14m': '/kaggle/input/pythia-output/pythia_analysis_20250327_210959/pythia_14m_data.json',
    '70m': '/kaggle/input/pythia-output/pythia_analysis_20250327_210959/pythia_70m_data.json',
    '160m': '/kaggle/input/pythia-output/pythia_analysis_20250327_210959/pythia_160m_data.json',
    '410m': '/kaggle/input/pythia-output/pythia_analysis_20250327_210959/pythia_410m_data.json',
    '1b': '/kaggle/input/pythia1b/pythia_1b_data.json',
    '1.4b': '/kaggle/input/pythia1b/pythia_1.4b_data.json'
}

step_display_names = {
    'pre_attention': 'Pre-Attention',
    'attention_patterns': 'Attention Patterns',
    'post_attention': 'Post-Attention',
    'post_attention_residual': 'Post-Attention Residual',
    'post_mlp': 'Post-MLP',
    'post_mlp_residual': 'Post-MLP Residual'
}

# Load all results
all_results = {}
for model_size, path in json_paths.items():
    with open(path, 'r') as f:
        all_results[model_size] = json.load(f)

# Steps we want to plot
steps = ['pre_attention', 'attention_patterns', 'post_attention', 
         'post_attention_residual', 'post_mlp', 'post_mlp_residual']

# Create one plot for each step
for step in steps:
    plt.figure(figsize=(15, 10))
    
    # Use a wider figure to accommodate legend
    colors = sns.color_palette("husl", len(all_results))
    
    # Plot each model's data
    for (model_size, color) in zip(all_results.keys(), colors):
        data = all_results[model_size]
        means = data['mean_ranks'][step]
        num_layers = data['num_layers']
        # Convert layer indices to percentages, using num_layers-1 as denominator 
        # since layer indices are 0-based
        layer_percentages = [(i/(num_layers-1)) * 100 for i in range(num_layers)]
        
        plt.plot(layer_percentages, means, 
                label=f'Pythia-{model_size}', 
                color=color, 
                marker='o',
                markersize=6,
                linewidth=2.5,
                markeredgewidth=2,
                markeredgecolor='white')
    
    plt.xlabel('Network Depth (%)', fontsize=14, labelpad=10)
    plt.ylabel('Effective Rank', fontsize=14, labelpad=10)
    plt.title(f'Effective Rank Across Layers:\n{step_display_names[step]}', 
              fontsize=16, pad=20)
    
    # Customize the grid
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Customize the legend
    plt.legend(bbox_to_anchor=(1.05, 1),
              loc='upper left',
              fontsize=12,
              frameon=True,
              fancybox=True,
              shadow=True)
    
    # Customize ticks
    plt.tick_params(axis='both', which='major', labelsize=12)
    
    # Add some padding to prevent cutoff
    plt.margins(x=0.02)
    
    # Make sure nothing gets cut off
    plt.tight_layout()
    
    # Save with high quality
    plt.savefig(f'{step}_comparison_new_.png', 
                dpi=300,
                bbox_inches='tight',
                facecolor='white',
                edgecolor='none')
    plt.close()

print("Comparison plots generated!")
```
